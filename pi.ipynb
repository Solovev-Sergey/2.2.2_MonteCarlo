```python
import numpy as np

def pi_error(N):
    data = np.random.rand(N,2)
    shift = np.array([[0.5, 0.5]]*N)
    data = data - shift
    data_of_len = np.array([(i[0]*i[0]+i[1]*i[1])**0.5 for i in data])
    data_of_len.sort()
    k = np.sum(data_of_len < 0.5)
    return (np.abs(k/N*4 - np.pi)) 
```

```python
import matplotlib.pyplot as plt
xs = [2**k for k in range(10, 25)]
ys = [abs(pi_error(x)) for x in xs]
plt.plot(xs, ys)
plt.xscale("log")
plt.yscale("log")
plt.show()
```
