```python
import numpy as np 
import matplotlib.pyplot as plt
import math
def pi_error(n):
    p = np.random.random((n, 2))  
    i = (p**2).sum(axis = 1) < 1.0 
    return 4*i.sum()/n - math.pi
xs = [2**k for k in range(10, 25)]
ys = [abs(pi_error(x)) for x in xs]
plt.plot(xs, ys)
plt.xscale("log")
plt.yscale("log")
plt.show()
```
